```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model
```

```python
# Load the model
model = load_model("waste_classification_model.keras")
```

```python
# Path to the test directory
test_dir = "DATASET/TEST"
```

```python
# Preprocessing test images
test_datagen = ImageDataGenerator(rescale=1.0/255.0)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=1,  # Process one image at a time
    class_mode='binary',
    shuffle=False  # Important to match the labels
)
```

```python
# Retrieve the actual labels
real_labels = test_generator.classes
label_map = {v: k for k, v in test_generator.class_indices.items()}  # Map index -> label
```

```python
# Make predictions
predictions = model.predict(test_generator, steps=len(test_generator), verbose=1)
predicted_labels = (predictions > 0.5).astype(int).flatten()  # 0 or 1
```

```python
# Count real and predicted categories
from collections import Counter

real_counts = Counter(real_labels)
predicted_counts = Counter(predicted_labels)
```

```python
# Format data for the graph
categories = ["Organic", "Recyclable"]
real_counts_formatted = [real_counts.get(0, 0), real_counts.get(1, 0)]
predicted_counts_formatted = [predicted_counts.get(0, 0), predicted_counts.get(1, 0)]
```

```python
# Create the graph
x = np.arange(len(categories))
width = 0.35  # Width of the bars

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, real_counts_formatted, width, label='Actual', color='skyblue')
plt.bar(x + width/2, predicted_counts_formatted, width, label='Predicted', color='orange')

# Configure axes
plt.xlabel('Categories')
plt.ylabel('Number of images')
plt.title('Comparison between actual and predicted categories')
plt.xticks(x, categories)
plt.legend()

# Display the graph
plt.show()
```
